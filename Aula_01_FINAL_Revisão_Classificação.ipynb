Pré Processamento

```python
# Conversão de uma coluna de dados categóricos para dados numéricos
dados['cut_encoded'] = LabelEncoder().fit_transform(dados['cut'])
dados.head()
```

```python
# Separação dos dados de entrada e dados de saída usando método .iloc

x = dados.iloc[:,-4:-1].values
x
#Dados de entrada --> x --> Features ---: Variáveis independentes
y = dados['cut_encoded'].values
y
```

```python
# Separação em dados de treino e dados de teste
#SEED
x_train, x_text, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
x_train
```

```python
# Tamanho do conjunto de treino
y_train.shape
```

```python
# Tamanho do conjunto de teste
y_test.shape
```

Escolha do algoritmo

```python
# Instanciar o modelo
modeloKNN = KNeighborsClassifier()
```

```python
# Treinar o modelo
MODELO = modeloKNN.fit(x_train, y_train)
```

```python
#gerar as previsões
y_predict = modeloKNN.predict(x_text)
y_predict
```

```python
acuracia = 100* (accuracy_score(y_test, y_predict))
print(f'A acurácia para do modelo baseado no algoritmo KNN é de: {acuracia: .2f} %')
```
